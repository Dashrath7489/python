```python
def Gender_encode(x):
    if x == "Male":
        return "M"
    else:
        return "F"
    
```

```python
mall["Gender_Category"] = mall["Gender"].apply(Gender_encode)
```

```python
def Income_Category(x):
    if x>0 and x<30:
        return "Lower Class"
    elif x>=30 and x<30:
        return "Middel Class"
    else:
        return "High Class"
```

```python
mall["IncomeCategory"] = mall["Annual Income"].apply(Income_Category)
```

```python
df1 = pd.DataFrame({"1key":["foo","bazz","reg","veg"], "value":[1,2,3,4]})
df2 = pd.DataFrame({"2key":["foo","bazz","reg","veg"], "value":[1,6,7,4]})
```

```python
df1,df2
```

```python
df1.merge(df2,on = "value",how = "inner")
```

```python
df1.merge(df2,on = "value",how = "outer")
```

```python
df1.merge(df2,on = "value",how = "left")
```

```python
df1.merge(df2,on = "value",how = "right")
```

```python
df1 = pd.DataFrame({"1key":["foo","bazz","reg","veg"], "value1":[1,2,3,4]})
df2 = pd.DataFrame({"2key":["foo","bazz","reg","veg"], "value2":[1,6,7,4]})
```

```python
df1,df2
```

```python
df1.join(df2.set_index("2key"), on="1key", how="left")
```

```python
df1.join(df2.set_index("2key"), on="1key", how="right")
```

```python
df1.join(df2.set_index("2key"), on="1key", how="inner")
```
