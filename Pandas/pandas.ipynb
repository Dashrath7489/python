```python
# float_col = ['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']
# for i in float_col:
#     df[i] = df[i].fillna(df[i].median())
float_col = ['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']
for i in float_col:
    df[i] = df[i].fillna(df[i].median())
```

```python
df["sex"] = df["sex"].fillna(df["sex"].mode()[0],inplace=True)
```

```python
df["bill_length_mm"] = pd.to_numeric(df["bill_length_mm"], errors="coerce")
```

```python
spec_group_df = (
    df.groupby(["species", "sex"])["bill_length_mm"]
      .mean()
)

spec_group_df

```

```python
spec_group_df = (
    df.groupby(["species", "sex"])["bill_length_mm"]
      .sum()
)

spec_group_df
```
