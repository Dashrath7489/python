# function

```python
def fun(a,b,c):
   sum = a + b + c
   return sum
print(fun(23,45,67))
```

```python
def func_name(*args):
    total = 0
    for i in args:
        total += i
    return total
# var = [1,2,3,4]
print(func_name(3,5,46,6,7))
```

```python
def k_wargs(**kwargs):
    for i, j in kwargs.items():
        print(i,j)
k_wargs(name = "dash",age = 21)
```

```python
def count(x):
    if x<= 1:
        return 1
    else:
        return x + count(x-1)
count(10)
```

```python
def fact(x):
    if x<= 1:
        return 1
    else:
        return x * fact(x-1)
fact(7)
```

```python
# fabonacci seriese using fanction

def Fab(n):
    a = 0
    b = 1
    for i in range(n):
        print(a)
        a,b = b, a+b
Fab(10)
```

```python
# fabonacci seriese using recursion

def fib(n):
    if n == 0:
        return 0
    if n == 1:
        return 1
    return fib(n-1) + fib(n-2)

for i in range(6):
    print(fib(i), end=" ")
```

```python
#filter method hai ye

ml = list(range(1,11))
print(ml)
nl = list(filter(lambda x : x > 5,ml))
print(nl)
```

```python
ml = list(range(1,11))
print(ml)
nl = list(filter(lambda x : x % 2 == 0,ml))
print(nl)
```

```python
ml = list(range(1,11))
print(ml)
nl = list(filter(lambda x : x % 2 != 0,ml))
print(nl)
```

```python
from functools import reduce
ml = list(range(1,11))
print(reduce(lambda x,y: x*y,ml))
```
