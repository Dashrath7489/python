```python
arr = np.array([[11,2,3],[56,570,58]])
print(arr)
two_arr_sort = np.sort(arr)
print(two_arr_sort)
```

```python
arr = np.array([[11,2,3],[56,570,58]])
print(arr)
two_arr_sort = np.sort(arr,axis=None)
print(two_arr_sort)
```

```python
arr1 = np.array([11,2,3])
arr2 = np.array([3,2,5])
print(arr1)
print(arr2)
```

```python
stack = np.stack((arr1,arr2),axis = 0)
print(stack)
```

```python
stack = np.stack((arr1,arr2),axis = 1)
print(stack)
```
