```python
import re
x = "machine learning"
pattern = "ac"
result = re.search(pattern,x)
print(result)
print(result.start())
print(result.end())
print(result.span())
```

```python
var = "hello 234 python  435 dfthr 4352"
pattern = "\d"
result = re.findall(pattern,var)
print(result)
```

```python
var = "hello 234 python  435 dfthr 4352"
pattern = "\D+"
result = re.findall(pattern,var)
print(result)
```

```python
import re
var = "hello 234 & python $ hello 435 dfhk 3453 "

pattern = "hello"
res = re.sub(pattern,"hii",var,count = 2)
print(res)
```

```python
import warnings
warnings.filterwarnings("ignore")
var = "hello 234 & python $ hello 435 dfhk 3453 "
pattern= "\d+"
result = re.finditer(pattern,var)

for i in result:
    print(i.span(),i.group())
```

```python
import re
import warnings
warnings.filterwarnings("ignore")

var = "hello 234 & python $ hello 435 dfhk 3453 "
pattern = "\d+"

result = re.split(pattern, var)
print(result)
```
