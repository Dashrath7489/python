```python
# 10. Calculate average excluding highest and lowest value. 
# Input: [10, 20, 30, 40, 50] 
# Output: 30.0

num = [10,20,30,40,50]

highest = num[0]
lowest = num[0]


for i in num:
    if i > highest:
        highest = i
    if i < lowest:
        lowest = i

avg = (highest + lowest) //2

print(avg)
```

```python
# 15. Sort list without using sort() or sorted(). 
# Input: [5, 2, 8, 1] 
# Output: [1, 2, 5, 8]

a = [5, 2, 8, 1]
for i in range(len(a)):
    for j in range(i+1, len(a)):
        if a[i] > a[j]:
            a[i], a[j] = a[j], a[i]
print(a)

```

```python
# 28. Replace duplicate values with -1. 
# Input: [1,2,2,3] 
# Output: [1,2,-1,3]
chars = [1,2,2,3]
seen = set()
result = []
for i in chars:
    if i in seen:
        result.append(-1)
    else:
        result.append(i)
        seen.add(i)
print(result)
```

```python
# 35. Find minimum and maximum element without using min(), max(). 
# Input: [10, 25, 5, 70] 
# Output: 5, 70 

num = [10, 25, 5, 70]

min_n = num[0]
max_n = num[0]

for i in num:
    if min_n > i:
        min_n = i

    if max_n < i:
        max_n = i

print(min_n,max_n)
    
```

```python
# 40. Sum of largest 2 numbers. 
# Input: [5, 20, 12, 8] 
# Output: 32 

lst = [5,20,12,8]

larg = []

for i in lst:
    if len(larg) < 2:
        larg.append(i)
    else:
        if i > min(larg):
            larg.remove(min(larg))
            larg.append(i)
print(sum(larg))


```

```python
# 41. Remove all occurrences of a particular element. 
# Input: [2, 3, 4, 2, 5] (Remove 2) 
# Output: [3, 4, 5]

lst = [2,3,4,2,5]

upd = []

for i in range(len(lst)):
    if lst[i] == lst[i]:
      lst.remove(i)
    else:
        upd.append(i)
print(upd)
```
