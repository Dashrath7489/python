```python
var = input("Enter your number: ")

if var == var[::-1]:
    print("palindrome",var)
else:
    print("not palindrome")
```

```python
n = input()
rev = ""

for ch in n:
    rev = ch + rev
    print(rev)

if rev == n:
    print("palindrome")

else:
    print("not palindrome")
```
