```python
password = ""
while password != "root":
    password = input("enter your password")

    if password != "root":
        print("invalid pass")

print("login")
```

```python
attempts = 3

while attempts != 0:
    password = input("enter password: ")

    if password == "root":
        print("login")
    else:
        attempts -= 1
        print("wrong pass")
if attempts == 0:
    print("please enter valid password")
```

# List Comprehension

```python
lst = [3,6,4,7,8,11,13,71,1,7,22]
lst_comp = [i**2 for i in lst]
print(lst_comp)
```

```python
lst = [3,6,4,7,8,11,13,71,1,7,22]
lst_comp = [i**2 for i in lst if i%2 != 0]
print(lst_comp)
```

```python
m = ["machine",2,5,54,67,75,34,65,"learning"]
l1 = [i for i in m if type(i)==str]
l2 = [i for i in m if type(i)==int]
print(l1)
print(l2)
```
