# Encapsulation

```python
class public():
    def _Show(self):
        print("public class")
pub = public()
pub._Show()
```

```python
class public():
    def _show(self):
        print("public Class")
class protected(public):
    def protect(self):
        print("This is protected method")
pub = protected()
pub._show()
```

```python
class Protect():
    def _protect(self):
        print("public class")
class public(protected):
    def protect(self):
        print("this is a protected method")
class Private():
    def __pri(self):
        print("private method")
Pvt = Private
```

```python
def Generator(x,y):
    sum = x+y
    yield sum
Gen = Generator(4,5)
for i in Gen:
    print(i)
```
