# dictionary


```python
a = {"name":"Dashrath","age":25,"city":"indore"}
print(type(a))
print(len(a))
print(a.keys())
print(a.values())
print(a.items())
print(a.get("name"))
print(a.get("city"))
```

```python
dic = {"name":["das","surendra","abhi","rahul"],"number":[34,56,78,43,]}

print(dic.get("number"))
```

```python
a = {"name":"Dashrath"}

b = {"city":"Indore"}

c = {"phone": 3456466746}

a.update(b)
a.update(c)
print(a)
```

```python
n1 = ["a","b","c"]
n2 = ["Das","rahul","virat"]

new_n1 = set(zip(n1,n2))
print(new_n1)
```

```python
n1 = ["a","b","c"]
n2 = ["Das","rahul","virat"]

new_n1 = dict(zip(n1,n2))
print(new_n1)
```

```python
n1 = ["a","b","c"]
n2 = ["Das","rahul","virat"]

new_n1 = list(zip(n1,n2))
print(new_n1)
```
