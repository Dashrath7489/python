```python
li = [1,3,2,4,6,7,9,4,6,5]
even = []
odd = []

for i in li:
    if i % 2 == 0:
        even.append(i)
    else:
        odd.append(i)
print(even)
print(odd)
```

```python
li = [1,3,5,7,9,10,15,20,34,88,100]
print(li)
new_li = []

for i in li:
    if i % 2 == 0 and i % 5 == 0:
        new_li.append(i)
print(new_li)
```

```python
# extract prime number from the list
 
li = [1,2,3,4,5,6,7,5,4,12,15,22,43,54]
prime = []
for i in li:
    if i>1:
        for j in range(2,i):
            if  i % j == 0:
                break 
        else:
            prime.append(i)
print(prime)
```
