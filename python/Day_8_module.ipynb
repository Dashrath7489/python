# Random

```python
import random

print(random.random())
```

```python


print(random.randint(1, 10))

names = ["Ram", "Shyam", "Mohan"]
print(random.choice(names))
```

```python
print(random.randrange(10,100,10))

print(random.choice(names,k = 2))
```

```python
for i in range(6):
    random.seed(1)
    print(random.randint(1,100))
```

```python
random.seed(9)
print(random.sample(range(10,100),10))
```

```python
var = calendar.Calendar(firstweekday=0)
for i in var.iterweekdays():
    print(i)
```
