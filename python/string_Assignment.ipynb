```python
# 1️ Question: Count the number of vowels in a given string. Input: "Hello World"

s = "Hello,World" 
count = 0

for i in s:
    if i in "aeiou":
        # print(i)
        count += 1
print(count)

```

```python
# 2️) Question: Reverse a given string. Input: "Python" Output: "nohtyP"

st = "python"
rev = ""

for i in st:
    rev = i + rev
print(rev)
```

```python
# 3️) Question: Check if a string is a palindrome.
#  Input: "madam" Output: True

a = input()

if a == a[::-1]:
    print("palindrom")
else:
    print("not palindrom")
```

```python
# 5) Question: Count the number of words in a string. 
# Input: "I love Python programming" 
# Output: 4 

string = "I love Python programming"
count = 1

for i in string:
    if i == " ":
        count+=1
print(count)
```

```python
# 6) Question: Find the frequency of each character in a string. 
# Input: "apple" 
# Output: {'a':1, 'p':2, 'l':1, 'e':1}

str1 = "apple"
count = {}

for i in str1:
    if i in count:
        count[i] += 1
    else:
        count[i] = 1
print(count)
```

```python
# 7) Question: Find all positions of a particular character in a string. 
# Input: "programming", character: 'g' 
# Output: [3, 10] 

a = "programming"
l = []
for i in range(len(a)):
    if a[i] == "g":
        l.append(i)
print(l)
```

```python
# 17) Question: Remove all special characters. 
# Input: "hello@123!" 
# Output: "hello123"

w = "hello@123!"
new_a = ""

for i in w:
    if i  not in "!@#$%^&*()":
        new_a += i
print(new_a)
```

```python
# 18) Question: Replace all vowels with *. 
# Input: "hello" 
# Output: "h*ll*" 

a = "hello"
vowel = "" 

for i in a:
    if i in "aeiou":
        vowel += "*"
    else:
        vowel += i
print(vowel)
```

```python
# 24) Question: Print characters at even positions. 
# Input: "Python" 
# Output: ['P', 't', 'o']

ch = "python"
new_ch = []

for i in range(len(ch)):
    if i%2 == 0:
        new_ch.append(ch[i])
print(new_ch)
```

```python
# 25) Question: Print characters at odd positions. 
# Input: "Python" 
# Output: ['y', 'h', 'n'] 

new = []
for i in range(len(ch)):
    if i%2 != 0:
        new.append(ch[i])
print(new)
```
